```python
import os
import numpy as np
import pandas as pd
```

```python
# Set directory and file names
data_dir = r'/Users/yasuyukiakita/Data'
results_dir = r'/Users/yasuyukiakita/Projects/python_gis_py312/results'
landfire_dir = os.path.join(data_dir, 'landfire')
lf_2023_fccs_csv = os.path.join(landfire_dir, 'LF2023_FCCS_240_CONUS', 
                                'CSV_Data', 'LF23_FCCS_240.csv')
```

```python
lf_csv_df = pd.read_csv(lf_2023_fccs_csv)
```

```python
lf_csv_df.dtypes
```

```python
lf_csv_df.head()
```

```python
lf_csv_base_df = lf_csv_df.loc[lf_csv_df['FCCSID'].str.isdigit()].reset_index(drop=True)
```

```python
lf_csv_base_df
```

```python
lf_csv_extended_df = lf_csv_df.loc[~lf_csv_df['FCCSID'].str.isdigit()].reset_index(drop=True)
```

```python
lf_csv_nodata_df = lf_csv_extended_df.loc[lf_csv_extended_df['VALUE']<0].reset_index(drop=True)
```

```python
lf_csv_nodata_df
```

```python
lf_csv_extended_df = lf_csv_extended_df.loc[lf_csv_extended_df['VALUE']>=0].reset_index(drop=True)
```

```python
lf_csv_extended_df
```

```python
lf_csv_extended_df['FCCSID_base'] = lf_csv_extended_df['FCCSID'].str.split('-', n=1).str[0].astype('int')
```

```python
lf_csv_extended_df
```

```python
lf_csv_base_df['FCCSID_base'] = lf_csv_base_df['FCCSID'].astype('int')
```

```python
lf_csv_base_df
```

```python
lf_csv_nodata_df['FCCSID_base'] = lf_csv_nodata_df['FCCS']
```

```python
lf_csv_nodata_df
```

```python
subclass_by_FCCSID_base_df = lf_csv_extended_df.groupby(
    'FCCSID_base', as_index=False).agg(func={'FCCS_REG':['min', 'max'], 
                                             'FUELBED':['min', 'max']})
```

```python
subclass_by_FCCSID_base_df.columns = ['FCCSID_base', 'FCCS_REG_min', 'FCCS_REG_max', 
                                      'FUELBED_min', 'FUELBED_max']
```

```python
subclass_by_FCCSID_base_df.loc[subclass_by_FCCSID_base_df['FCCS_REG_min'] != subclass_by_FCCSID_base_df['FCCS_REG_max']]
```

```python
subclass_by_FCCSID_base_df.loc[subclass_by_FCCSID_base_df['FUELBED_min'] != subclass_by_FCCSID_base_df['FUELBED_max']]
```

```python
lf_csv_final_df = pd.concat([lf_csv_nodata_df, lf_csv_base_df, lf_csv_extended_df], ignore_index=True)
```

```python
assert lf_csv_final_df.shape[0] == lf_csv_df.shape[0], 'lf_csv_final_df error'
```

```python
lf_csv_final_df = lf_csv_final_df.loc[:, ['VALUE', 'FCCSID_base', 'FCCS', 'FCCSID', 'FCCS_REG', 'FUELBED']]
```

```python
fccsid_reg_fuelbd_cnt_df = lf_csv_final_df.groupby(['FCCSID_base', 'FCCS_REG', 'FUELBED'], as_index=False).size()
```

```python
fccsid_cnt_df = fccsid_reg_fuelbd_cnt_df.groupby(['FCCSID_base'], as_index=False).size()
```
